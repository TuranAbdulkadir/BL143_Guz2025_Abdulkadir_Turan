```python
print("--- Akıllı Gün Planlayıcıya Hoş Geldiniz ---")
print("Lütfen süreleri saat cinsinden giriniz (Örn: 7.5 veya 8)\n")

while True:
    try:
        uyku_suresi = float(input("Günlük uyku süreniz (saat): "))
        ders_suresi = float(input("Günlük ders çalışma süreniz (saat): "))
        serbest_zaman = float(input("Günlük serbest zamanınız (saat): "))

        if uyku_suresi < 0 or ders_suresi < 0 or serbest_zaman < 0:
            print("HATA: Süreler negatif olamaz! Lütfen tekrar giriniz.\n")

        toplam_sure = uyku_suresi + ders_suresi + serbest_zaman

        if toplam_sure > 24:
            print(f"HATA: Girdiğiniz sürelerin toplamı ({toplam_sure} saat) 24 saati geçiyor!")
            print("Lütfen bir güne sığacak mantıklı değerler giriniz.\n")
            continue

        break

    except ValueError:
        print("HATA: Lütfen sadece sayısal değer giriniz (Örn: 5 veya 6.5)!\n")


if uyku_suresi < 6:
    uyku_durumu = "Yetersiz uyku"
elif 6 <= uyku_suresi <= 8:
    uyku_durumu = "Normal uyku"
else:
    uyku_durumu = "Fazla uyku"


if ders_suresi >= 6:
    calisma_durumu = "Yoğun çalışma günü"
else:
    calisma_durumu = "Normal çalışma günü"


if uyku_suresi < 6 and ders_suresi >= 6:
    genel_degerlendirme = "Verimsiz gün"
else:
    genel_degerlendirme = "Dengeli gün"

print("\n--- DEĞERLENDİRME SONUCU ---")
print(f"Uyku durumu: {uyku_durumu}")
print(f"Çalışma durumu: {calisma_durumu}")
print(f"Genel değerlendirme: {genel_degerlendirme}")
```
